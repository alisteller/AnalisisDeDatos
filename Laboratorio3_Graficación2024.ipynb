```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("casasboston.csv")
```

```python
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python

df.N_HABITACIONES_MEDIO.plot.hist(edgecolor='white',color="lightblue")
plt.xlabel("Número de habitaciones")
plt.ylabel("Frecuencia")
plt.title("Promedio del número de habitaciones ")
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=1,s=30, edgecolor="white",color="red")
plt.ylabel("Índice de crimen")
plt.xlabel("Promedio")
plt.title("Representación del índice de crimen y su promedio")
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
```

```python
colors= ['#B298DC','pink','lightblue','#C1EBE9']
valor_por_ciudad.head(10).plot.barh(width=0.9,color=colors)
plt.ylabel("Ciudad")
plt.xlabel("Promedio")
plt.title("Representación de crimenes por ciudad")
plt.show()
```
